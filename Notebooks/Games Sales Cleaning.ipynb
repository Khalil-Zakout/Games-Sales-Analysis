```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"../Excel Data/Games Sales.csv")
df = df.drop_duplicates()
df
```

```python
# Deleting Unnecessary Columm: 

df = df.drop(columns="img")
df = df.drop(columns= "last_update")
```

```python
# Deleting Games Published Before 2013:

df["release_date"] = pd.to_datetime(df["release_date"])
df = df[df["release_date"] >= pd.Timestamp('2013-01-01')]
df.reset_index(drop=True, inplace=True)
df
```

```python
# Drop Missing Values to Work With Complete Data:
 
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
# Changing Number of Sales into Millions:

df[["total_sales","na_sales","jp_sales","pal_sales","other_sales"]] = df[["total_sales","na_sales","jp_sales","pal_sales","other_sales"]].apply(lambda x: x * 1000000).round(2)
df
```

```python
# Saving The Cleaned Dataset:

df.to_excel(r"../Excel Data/Cleaned Games Sales.xlsx",index=False)
print("Success")
```
